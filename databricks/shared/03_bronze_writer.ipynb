```python
from pyspark.sql.functions import current_timestamp


def start_bronze_writer(
    stream_df,
):

    bronze_df = (

        stream_df

            .withColumn(
                "ingestion_timestamp",
                current_timestamp(),
            )
    )

    return (

        bronze_df.writeStream

            .format("delta")

            .outputMode("append")

            .option(
                "checkpointLocation",
                BRONZE_CHECKPOINT,
            )

            .toTable(
                BRONZE_TABLE,
            )
    )
```
